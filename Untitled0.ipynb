```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

print("환경 준비 완료!")
```

```python
import yfinance as yf

data = yf.download("AAPL", start="2020-01-01", progress=False)

data.head()
```

```python
data['Return'] = data['Close'].pct_change()
data.head()
```

```python
data['Target'] = (data['Return'].shift(-1) > 0).astype(int)
data.tail()
```

```python
data = data.dropna()
```

```python
features = ['Open','High','Low','Close','Volume']

x = data[features]
y = data['Target']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x,y,test_size = 0.2, shuffle = False
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

pred = model.predict(x_test)

accuracy = accuracy_score(y_test,pred)
print("Accuracy:", accuracy)
```

```python
#이동평균 추가 (지금부터 정확도 올리기 위해 하는 단계)
data['MA5'] = data['Close'].rolling(5).mean()
data['MA20'] = data['Close'].rolling(20).mean()
```

```python
#변동성 추가
data['Volatillity'] = data['Close'].rolling(5).std()
```

```python
#코드 정리
data = data.dropna()
```

```python
#새로운 feature만들기
features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA5', 'MA20', 'Volatility']

X = data[features]
y = data['Target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, shuffle=False
)

model = LogisticRegression()
model.fit(x_train, y_train)

pred = model.predict(x_test)
accuracy = accuracy_score(y_test, pred)

print("Accuracy:", accuracy)
#간단한 데이터이고 단순한 모델이라 주가 예측같은 문제는 거의 못 맞춤
```

```python
from sklearn.ensemble import RandomForestClassifier
#logistic regression은 직선모델 BUT, RandomForest는 복잡한 패턴을 잡음
model = RandomForestClassifier(n_estimators=100)
model.fit(x_train, y_train)

pred = model.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, pred)

print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, pred))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

plt.scatter(range(100), y_test.values[:100], label='Actual')
plt.scatter(range(100), pred[:100], label='Predicted')

plt.legend()
plt.title("Prediction Comparison")
plt.show()
```

```python
#주가 예측하기
#예측이 1이면 매수, 0이면 아무것도 안 함
returns = data['Return'].iloc[-len(y_test):] #테스트 구간 수익률 가져오기
```

```python
strategy_returns = returns * pred
```

```python
strategy_returns = returns * pred #전략 수익률

import numpy as np #누적 수익 계산

cumulative_returns = (1 + strategy_returns).cumprod()

market_returns = (1 + returns).cumprod() #그냥 주식이랑 비교

import matplotlib.pyplot as plt #그래프 비교

plt.figure(figsize=(12,5))

plt.plot(cumulative_returns.values, label='Strategy')
plt.plot(market_returns.values, label='Buy & Hold')

plt.legend()
plt.title("Strategy vs Market")
plt.show()
#그래프에서 전략 > 시장 이면 모델을 쓸 가치가 있음
#반대면 그냥 들고있는게 더 나음
```
